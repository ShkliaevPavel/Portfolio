Exercise 1

Write a function that classifies films from lesson materials according to the following rules:
- rating 2 and below - low rating;
- rating 4 and below - average rating;
- rating 4.5 and 5 - high rating.

Write the classification result in the "class" column.

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
def class_(rating):
    if rating <= 2:
      return 'low'
    elif rating <= 4:
      return 'middle'
    else:
      return 'high'

ratings['class'] = ratings['rating'].apply(class_)
ratings.head()
```

Use the keywords.csv file.

You need to write a geo-classifier that can assign each line a geographic affiliation to a specific region.
If a query contains the name of a city in a region, then the name of this region is written in the ‘region’ column. If the search query does not contain the name of the city, then set it to ‘undefined’.

Rules for distribution by regions Center, North-West and Far East:

Write the classification result in a separate region column.

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
keywords = pd.read_csv('keywords.csv')
keywords.head()
```

```python
def geo(keyword):
    for key, values in geo_data.items():
      for each_value in values:
        if each_value in keyword:
          return key
    return 'undefined'
```

```python
keywords['region'] = keywords['keyword'].apply(geo)
keywords.head()
```

There is an opinion that they used to make real movies, not like now. Your task is to check this statement using the files with movie ratings from the previous homework: the movies.csv file and ratings.csv from the database. We need to check whether it is true that as the year of release of a film increases, its average rating becomes lower.

You will not touch upon the subjective factors of issuing these ratings, but will go through the algorithm:

In the years variable, write a list of all years from 1950 to 2010.

Write a function production_year that sets the release year for each line of the movie title.

Not all movie titles contain the release year in the same format, so use the following algorithm:

for each line, go through all the years in the years list;

* if the year number is present in the movie title, then the function returns this year as the year of release;
* if none of the year numbers from the years list are found in the film title, then 1900 is returned.

Write down the year of release of the film according to the algorithm in step 2 in the new column ‘year’.

Calculate the average rating of all films for each value in the ‘year’ column and sort the result in descending order of rating.

```python
movies.head()
```

```python
def production_year(title):
    for i in years:
      if  str(i) in title:
        return i
    return 1900
```

```python
years = list(range(1950,2011))
filtered_movies = movies.filter(items = ['movieId','title'])
filtered_ratings = ratings.filter(items = ['movieId', 'rating'])
merged = filtered_movies.merge(filtered_ratings, on= 'movieId')
merged['year'] = merged['title'].apply(production_year)
merged.groupby('year')[['rating']].mean().sort_values('rating', ascending = False)
```
