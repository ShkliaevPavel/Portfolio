Exercise 1.

For the log dataframe from the lesson materials, create a source_type column according to the rules:

1. if the traffic_source is "Yandex" or "Google", then the source_type is "organi"c;
2. for "paid" and "email" sources from Russia, set "ad";
3. for "paid" and "email" sources not from Russia, set "other";
4. all other options from traffic_source must stay without changes.






```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv', sep = ';')
log.head()
```

```python
def source_type(log):
    if log['traffic_source'] in ('yandex','google'):
      return 'organic'
    elif log['traffic_source'] == 'paid' and log['region'] == 'Russia':
      return 'ad'
    elif log['traffic_source'] == 'paid':
      return 'other'
    return log['traffic_source']
```

```python
log['source_type'] = log.apply(source_type, axis = 1)
log.head(10)
```

Task 3

Use the movie ratings file ml-latest-small/ratings.csv.

Calculate the average lifetime of users who have given more than 100 ratings.

Lifetime refers to the difference between the maximum and minimum values of the timestamp column for a given userId value.

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
group = ratings.groupby('userId').agg({'timestamp': ['min', 'max', 'count']}) #Группировка для подсчета кол-ва оценок каждого пользователя
filtered_group = group.loc[group['timestamp']['count'] >100] #Выборка пользователей с кол-вом оценок больше 100
mean = (filtered_group[('timestamp','max')] - filtered_group[('timestamp','min')]).mean() #Расчет среднего времени жизни
mean
```

Statistics of the company's customer transportation services by type are given.

You need to create two tables:
* a table with three types of revenue for each client_id without specifying the client's address;
* a similar table by type of revenue indicating the client’s address.





```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1',
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
from functools import reduce
```

```python
reduce(lambda a, b: a.merge(b, on = 'client_id', how = 'outer'), [rzd, auto, air])
```

```python
(client_base
 .merge(rzd, how = 'left', left_on = 'client_id', right_on = 'client_id')
 .merge(auto, how = 'left', left_on = 'client_id', right_on = 'client_id')
 .merge(air, how = 'left', left_on = 'client_id', right_on = 'client_id'))
```
