Determine which movie received the most 5.0 ratings.

```python
import pandas as pd
```

Task 2

Using the power.csv file, calculate the total consumption of the Baltic countries (Latvia, Lithuania and Estonia) of categories 4, 12 and 21 for the period from 2005 to 2010.
Do not take negative quantity values into account in your calculations.

```python
power = pd.read_csv('power.csv')
```

```python
def baltic(country):
  if country in ['Latvia', 'Lithuania', 'Estonia']:
    return 'Baltic region'
  return 'Other'

power['Region'] = power.country.apply(baltic)
period = pd.Series(range(2005,2011))
categories = (4,12,21)

filtered = power[(power['Region']=='Baltic region' ) &\
                 (power['year'].isin(period)) &\
                 (power['category'].isin(categories)) &\
                 (power['quantity'] >0)]

print(sum(filtered['quantity']))
```
